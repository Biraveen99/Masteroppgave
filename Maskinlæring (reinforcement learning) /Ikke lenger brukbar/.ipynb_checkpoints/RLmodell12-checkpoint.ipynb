```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Last inn CSV-filen (erstatt med din filbane)
file_path = '/Users/biraveennedunchelian/Documents/Masteroppgave/Masteroppgave/Newest addition/Gausian curve/gaussian_curve_fixedidk.csv'
df = pd.read_csv(file_path)

# En enkel RL-inspirert modell som bruker et fast vindu og forsterker seg ved  justere vekter basert p feil
class SimpleRLModel:
    def __init__(self, window_size=10, learning_rate=0.01):
        self.window_size = window_size
        self.learning_rate = learning_rate
        self.weights = np.random.rand(window_size)

    def predict(self, window):
        return np.dot(self.weights, window)

    def update(self, window, true_value):
        prediction = self.predict(window)
        error = true_value - prediction
        self.weights += self.learning_rate * error * window
        return prediction

# Glatt dataene ved  bruke et glidende vindu (f.eks. 10 tidssteg)
window_size_smooth = 10
smoothed_data = np.convolve(df['Y'].values, np.ones(window_size_smooth)/window_size_smooth, mode='valid')

# Del dataene i 80% for trening og 20% for testing
train_size = int(len(smoothed_data) * 0.8)
train_data = smoothed_data[:train_size]
test_data = smoothed_data[train_size:]

# Reinitialiser modellen
model = SimpleRLModel(window_size=10, learning_rate=0.0005)

# Tren modellen p de glatte treningsdataene
predictions_train = []
true_values_train = []

train_data_normalized = (train_data - np.min(train_data)) / (np.max(train_data) - np.min(train_data))

# Tren modellen
for i in range(10, len(train_data_normalized)):
    window = train_data_normalized[i - 10:i]
    true_value = train_data_normalized[i]
    prediction = model.update(window, true_value)
    predictions_train.append(prediction)
    true_values_train.append(true_value)

# Reskalere treningsprediksjonene tilbake til original skala
predictions_train_rescaled = np.array(predictions_train) * (np.max(df['Y'].values) - np.min(df['Y'].values)) + np.min(df['Y'].values)
true_values_train_rescaled = np.array(true_values_train) * (np.max(df['Y'].values) - np.min(df['Y'].values)) + np.min(df['Y'].values)

# Validering p de glatte testdataene
predictions_test = []
true_values_test = []

test_data_normalized = (test_data - np.min(test_data)) / (np.max(test_data) - np.min(test_data))

# Gjr prediksjoner for testdataene
for i in range(10, len(test_data_normalized)):
    window = test_data_normalized[i - 10:i]
    true_value = test_data_normalized[i]
    prediction = model.predict(window)
    predictions_test.append(prediction)
    true_values_test.append(true_value)

# Reskalere testprediksjonene tilbake til original skala
predictions_test_rescaled = np.array(predictions_test) * (np.max(df['Y'].values) - np.min(df['Y'].values)) + np.min(df['Y'].values)
true_values_test_rescaled = np.array(true_values_test) * (np.max(df['Y'].values) - np.min(df['Y'].values)) + np.min(df['Y'].values)

# Plot resultatene
plt.figure(figsize=(14, 6))

# Venstre: Prediksjon for treningsdataene
plt.subplot(1, 2, 1)
plt.plot(range(len(true_values_train_rescaled)), true_values_train_rescaled, label='Actual Training CPU Usage')
plt.plot(range(len(predictions_train_rescaled)), predictions_train_rescaled, label='Predicted Training CPU Usage', linestyle='--')
plt.title('Training Data: Actual vs Predicted')
plt.xlabel('Time Step')
plt.ylabel('CPU Usage')
plt.legend()
plt.grid(True)

# Hyre: Prediksjon for testdataene (validering)
plt.subplot(1, 2, 2)
plt.plot(range(len(true_values_test_rescaled)), true_values_test_rescaled, label='Actual Test CPU Usage')
plt.plot(range(len(predictions_test_rescaled)), predictions_test_rescaled, label='Predicted Test CPU Usage', linestyle='--')
plt.title('Test Data: Actual vs Predicted (Validation)')
plt.xlabel('Time Step')
plt.ylabel('CPU Usage')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
