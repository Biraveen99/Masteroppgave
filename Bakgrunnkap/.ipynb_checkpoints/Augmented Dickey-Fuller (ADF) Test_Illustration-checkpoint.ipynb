```python
import matplotlib.pyplot as plt
import pandas as pd

# Create sample time series data
time = list(range(1, 11))
y = [10, 12, 15, 14, 13, 17, 18, 20, 19, 21]
delta_y = [None] + [y[i] - y[i-1] for i in range(1, len(y))]

# Put into DataFrame
df = pd.DataFrame({
    'Time': time,
    'y_t': y,
    'Î”y_t': delta_y
})

# Plot with markers for both lines
plt.figure(figsize=(10, 5))
plt.plot(df['Time'], df['y_t'], marker='o', label='$y_t$ (Original Series)', color='blue')
plt.plot(df['Time'], df['Î”y_t'], marker='x', linestyle='--', label='$\Delta y_t$ (First Difference)', color='orange')
plt.scatter(df['Time'], df['Î”y_t'], color='orange', zorder=5)

plt.title('Time Series data and First Difference')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save the figure
plt.savefig("lagged_difference_plot.png", dpi=300)            
plt.show()
```
