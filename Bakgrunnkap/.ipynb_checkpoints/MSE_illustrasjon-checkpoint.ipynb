```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data points (X, Y) and predicted Y values (Y_hat)
X = np.array([1, 2, 3, 4, 5, 6, 7])
Y_actual = np.array([1, 2, 3, 4, 5, 6, 7])
Y_predicted = np.array([2, 3, 4, 4.5, 5, 5.5, 6])

# Plot actual points
plt.figure(figsize=(8, 6))
plt.scatter(X, Y_actual, color='purple', s=100, label='Actual (Y)')

# Plot prediction line
plt.plot(X, Y_predicted, color='blue', label='Prediction Line')

# Plot error lines
for x, y, y_hat in zip(X, Y_actual, Y_predicted):
    plt.plot([x, x], [y, y_hat], color='red')  # vertical error line

# Annotate points
for x, y in zip(X, Y_actual):
    plt.text(x - 0.25, y + 0.3, f"({x},{y})", fontsize=9)

# Labels and legend
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Illustration of MSE with Actual and Predicted Values")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the figure
plt.savefig("mse_illustration.png")
plt.show()
```
